```python
#opening webpage using chrome webdriver
driver = webdriver.Chrome()
driver.get('https://www.bikedekho.com/new-bikes#all_brands')
time.sleep(3)

#Imports the HTML of the current page into python
soup = BeautifulSoup(driver.page_source, 'lxml')
```

```python
#Extraching Link for each brand
postings = soup.find_all('li' , class_ = 'gsc_col-xs-4 gsc_col-sm-3 gsc_col-md-3 gsc_col-lg-2')
for post in postings:
    link = post.find('a' , class_ = 'BrIconNewCar').get('href')
    full_link = 'https://www.bikedekho.com'+link
    print(full_link)
```

```python
#1) Bike Name
postings = soup.find_all('li' , class_ = 'gsc_col-xs-12 gsc_col-sm-6 gsc_col-md-12 gsc_col-lg-12')

for post in postings:
    try:
        name = post.find('div' , class_ = 'gsc_col-sm-12 gsc_col-xs-12 gsc_col-md-8 listView holder posS').find('h3').text
        print(name)
    except:
        name = 'Advertisement'
        print(name)
```

```python
#2) Price
postings = soup.find_all('li' , class_ = 'gsc_col-xs-12 gsc_col-sm-6 gsc_col-md-12 gsc_col-lg-12')

for post in postings:
    try:
        price = post.find('div' , class_ = 'price').text
        print(price)
    except:
        price = 'Advertisement'
        print(price)
```

```python
#3) Milage
postings = soup.find_all('li' , class_ = 'gsc_col-xs-12 gsc_col-sm-6 gsc_col-md-12 gsc_col-lg-12')

for post in postings:
    milage = post.find(string = re.compile('kmpl'))
    print(milage)
```

```python
#4) Engine power(in PS)

postings = soup.find_all('li' , class_ = 'gsc_col-xs-12 gsc_col-sm-6 gsc_col-md-12 gsc_col-lg-12')

for post in postings:
    engine_power = post.find(string = re.compile('PS'))
    print(engine_power)
```

# Complete Code

```python
df = pd.DataFrame({'Brand':[''], 'Name':[''], 'Price':[''], 'Milage':[''], 'Engine Power':['']})
df
```

```python
#opening webpage using chrome webdriver
driver = webdriver.Chrome()
driver.get('https://www.bikedekho.com/new-bikes#all_brands')
time.sleep(3)

#Imports the HTML of the current page into python
soup = BeautifulSoup(driver.page_source, 'lxml')
#Extraching Link for each brand
url_list = []
brand_lis = []
postings1 = soup.find_all('li' , class_ = 'gsc_col-xs-4 gsc_col-sm-3 gsc_col-md-3 gsc_col-lg-2')
for post1 in postings1:
    link = post1.find('a' , class_ = 'BrIconNewCar').get('href')
    full_link = 'https://www.bikedekho.com'+link
    url_list.append(full_link)
    brand_ = post1.find('a' , class_ = 'BrIconNewCar').find('span').text
    brand_lis.append(brand_)

#opening webpage using chrome webdriver
cnt = 0
driver = webdriver.Chrome()
for url in url_list:
    driver.get(url)
    time.sleep(3)

    #scroll till end
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')

    #Imports the HTML of the current page into python
    soup = BeautifulSoup(driver.page_source, 'lxml')

    #Extracting Data
    postings = soup.find_all('li' , class_ = 'gsc_col-xs-12 gsc_col-sm-6 gsc_col-md-12 gsc_col-lg-12')
    for post in postings:
        #Brand
        brand = brand_lis[cnt]
        #extracting name of bike
        try:
            name = post.find('div' , class_ = 'gsc_col-sm-12 gsc_col-xs-12 gsc_col-md-8 listView holder posS').find('h3').text
        except:
            name = 'Advertisement'
        #extracting price
        try:
            price = post.find('div' , class_ = 'price').text
        except:
            price = 'Advertisement'
        #extracting milage
        milage = post.find(string = re.compile('kmpl'))
        #extracting engine power
        engine_power = post.find(string = re.compile('PS'))

        df = df.append({'Brand':brand, 'Name':name, 'Price':price, 'Milage':milage, 'Engine Power':engine_power} , ignore_index = True)
        
    cnt = cnt+1
```
